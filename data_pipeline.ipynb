```python
from pathlib import Path

base_path = Path("data/real")
base_path.mkdir(parents=True, exist_ok=True)

for dataset in selected_datasets:
    # nazwa folderu z label_name
    label_name = dataset[0]["label_name"]
    dataset_dir = base_path / label_name
    dataset_dir.mkdir(parents=True, exist_ok=True)

    for i, example in enumerate(dataset):
        image = example["image"]  # PIL.Image

        # JPEG-safe
        if image.mode != "RGB":
            image = image.convert("RGB")

        image.save(dataset_dir / f"{i:06d}.jpg", format="JPEG")

    print(f"‚úî {label_name}: zapisano {len(dataset)} obraz√≥w")
```

# Generate AI images of mushrooms

```python
import os

base_path = "data/real"
mushrooms = os.listdir(base_path)
mushrooms
```

```python
styles = [
    "macro photography",
    "natural forest light",
    "studio product photo on neutral background",
    "cinematic shallow depth of field",
    "documentary photo"
]
backgrounds = [
    "on moss in a forest",
    "on wet leaves after rain",
    "near tree roots with pine needles",
    "on dark soil with small twigs",
    "on a wooden log"
]
angles = [
    "low angle shot",
    "side view",
    "top-down view",
    "extreme macro close-up of gills",
    "close-up of the cap texture"
]
time_weather = [
    "golden hour light",
    "overcast soft light",
    "foggy morning",
    "after rain with water droplets",
    "sunlight through trees (bokeh)"
]
```

```python
def build_prompt(species: str) -> str:
    return (
        f"A realistic photo of {species} mushroom, "
        f"{random.choice(styles)}, "
        f"{random.choice(backgrounds)}, "
        f"{random.choice(angles)}, "
        f"{random.choice(time_weather)}, "
        "high detail, natural colors, sharp focus on the mushroom, "
        "no text, no watermark"
    )
```

```python
import base64

def save_b64_image(b64_json: str, path: Path):
    img_bytes = base64.b64decode(b64_json)
    path.write_bytes(img_bytes)
```

```python
from openai import OpenAI
from dotenv import load_dotenv
from tqdm import tqdm

load_dotenv()

client = OpenAI()

out_dir = Path("data/synth")
out_dir.mkdir(parents=True, exist_ok=True)

images_per_species = 23  # ile obraz√≥w na grzyba

for species in tqdm(mushrooms):
    species_dir = out_dir / species
    species_dir.mkdir(exist_ok=True)

    for i in range(images_per_species):
        prompt = build_prompt(species.replace("_", " "))

        result = client.images.generate(
            model="gpt-image-1-mini",
            prompt=prompt,
            size="1024x1024",
        )

        save_b64_image(result.data[0].b64_json, species_dir / f"{i+1:04d}.png")

print("Done.")

```

## Preparing test dataset

```python
from pathlib import Path
import shutil
from typing import Iterable, Union


def sample_from_roots(
    roots: Iterable[Union[str, Path]],
    n_per_class: int,
    dest_root: Union[str, Path],
    *,
    seed: int | None = None,
    allowed_ext: tuple[str, ...] = (".jpg", ".jpeg", ".png", ".bmp", ".webp"),
    mode: str = "copy",  # "copy" albo "move"
) -> dict[str, int]:
    """
    roots:
        lista ≈õcie≈ºek nadrzƒôdnych, np. ["data/real", "data/synth"]
        ka≈ºda zawiera podfoldery = klasy
    n_per_class:
        ile obraz√≥w losowaƒá z KA≈ªDEJ klasy
    dest_root:
        ≈õcie≈ºka docelowa (powstanƒÖ podfoldery klas)
    mode:
        "copy" ‚Üí kopiuj obrazy
        "move" ‚Üí przenie≈õ obrazy
    """

    if mode not in {"copy", "move"}:
        raise ValueError("mode musi byƒá 'copy' albo 'move'")

    rng = random.Random(seed)
    dest_root = Path(dest_root)
    dest_root.mkdir(parents=True, exist_ok=True)

    op = shutil.copy2 if mode == "copy" else shutil.move

    stats: dict[str, int] = {}

    for root in map(Path, roots):
        if not root.exists():
            raise FileNotFoundError(f"Nie istnieje: {root}")

        for class_dir in root.iterdir():
            if not class_dir.is_dir():
                continue

            class_name = class_dir.name
            out_dir = dest_root / class_name
            out_dir.mkdir(parents=True, exist_ok=True)

            images = [
                p for p in class_dir.iterdir()
                if p.is_file() and p.suffix.lower() in allowed_ext
            ]

            if not images:
                continue

            k = min(n_per_class, len(images))
            chosen = rng.sample(images, k)

            for img in chosen:
                op(img, out_dir / img.name)

            stats[class_name] = stats.get(class_name, 0) + len(chosen)

    return stats
```

## Preparing dataset for E1

```python
from pathlib import Path
from typing import Iterable, Union, Mapping


def sample_from_roots(
    roots: Iterable[Union[str, Path]],
    dest_root: Union[str, Path],
    *,
    n_per_root: Union[int, Mapping[Union[str, Path], int]] | None = None,
    seed: int | None = None,
    allowed_ext: tuple[str, ...] = (".jpg", ".jpeg", ".png", ".bmp", ".webp"),
    mode: str = "copy",      # "copy" | "move"
    take_all: bool = False,  # True ‚Üí kopiuj/przenie≈õ WSZYSTKO
) -> dict[str, int]:
    """
    roots:
        lista ≈õcie≈ºek nadrzƒôdnych (ka≈ºda zawiera foldery klas)
    dest_root:
        katalog docelowy
    n_per_root:
        - int ‚Üí tyle obraz√≥w z KA≈ªDEJ klasy w KA≈ªDYM root
        - dict[root] = int ‚Üí osobna liczba dla ka≈ºdej ≈õcie≈ºki
        - None ‚Üí wymagane, je≈õli take_all=True
    mode:
        "copy" albo "move"
    take_all:
        True ‚Üí bierze wszystkie obrazy (ignoruje n_per_root)
    """

    if mode not in {"copy", "move"}:
        raise ValueError("mode musi byƒá 'copy' albo 'move'")

    if not take_all and n_per_root is None:
        raise ValueError("Podaj n_per_root albo ustaw take_all=True")

    rng = random.Random(seed)
    dest_root = Path(dest_root)
    dest_root.mkdir(parents=True, exist_ok=True)

    op = shutil.copy2 if mode == "copy" else shutil.move

    stats: dict[str, int] = {}

    for root in map(Path, roots):
        if not root.exists():
            raise FileNotFoundError(f"Nie istnieje: {root}")

        # ile obraz√≥w z tego root
        if isinstance(n_per_root, dict):
            k_root = n_per_root.get(root) or n_per_root.get(str(root))
            if not take_all and k_root is None:
                raise ValueError(f"Brak liczby dla root: {root}")
        else:
            k_root = n_per_root

        for class_dir in root.iterdir():
            if not class_dir.is_dir():
                continue

            class_name = class_dir.name
            out_dir = dest_root / class_name
            out_dir.mkdir(parents=True, exist_ok=True)

            images = [
                p for p in class_dir.iterdir()
                if p.is_file() and p.suffix.lower() in allowed_ext
            ]

            if not images:
                continue

            if take_all:
                chosen = images
            else:
                k = min(k_root, len(images))
                chosen = rng.sample(images, k)

            for img in chosen:
                op(img, out_dir / img.name)

            stats[class_name] = stats.get(class_name, 0) + len(chosen)

    return stats
```

```python
sample_from_roots(
    ['data/real', 'data/synth'],
    'E1/train', 
    take_all = True,
)
```
